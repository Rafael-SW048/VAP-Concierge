```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_csv("./inferDiff-1.csv", index_col=False)
```

```python
def readFile(idx):
    infer_diff = []
    with open(f"./inferDiff-{idx}.csv", "r") as file:
        infer_diff = json.load(file)
    acc_sen = []
    with open(f"./accSen-{idx}.csv", "r") as file:
        acc_sen = json.load(file)
    return infer_diff[:-1], acc_sen[:-1]
```

```python
if1, as1 = readFile(1)
if2, as2 = readFile(2)
if3, as3 = readFile(3)
```

```python
plt.scatter(if1, as1, label="Jakarta, corr = 0.76")
plt.scatter(if2, as2, label="Miami, corr = 0.707")
plt.scatter(if3, as3, label="Coldwater, corr = 0.624")
plt.xlabel("InferDiff")
plt.ylabel("AccSen")
plt.title("InferDiff vs AccSen (Upwards)")
plt.legend(loc="lower right")
```
